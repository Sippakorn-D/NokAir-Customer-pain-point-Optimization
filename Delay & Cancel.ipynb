# The Poisson Process

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_delay = 19   # Average daily flight delays
lambda_cancel = 30  # Average daily flight cancellations
days = 7         # Number of days to simulate

# Calculate mean for the period
mu_delay = lambda_delay * days
mu_cancel = lambda_cancel * days

# Create range of k values
k_values_delay = np.arange(0, 700)
k_values_cancel = np.arange(0, 700)

# Calculate Poisson probabilities
prob_delay = poisson.pmf(k_values_delay, mu_delay)
prob_cancel = poisson.pmf(k_values_cancel, mu_cancel)
```

# The Result by Max Probability

```python
# Max probability of Delayed flights
max_prob_index_Delay = np.argmax(prob_delay)
max_prob_k_Delay = k_values_delay[max_prob_index_Delay]
print("Number of delayed flights with max probability:", max_prob_k_Delay)
print("Max probability value:", prob_delay[max_prob_index_Delay])

# Max probability of Canceled flights
max_prob_index_Cancel = np.argmax(prob_cancel)
max_prob_k_Cancel = k_values_cancel[max_prob_index_Cancel]
print("Number of canceled flights with max probability:", max_prob_k_Cancel)
print("Max probability value:", prob_cancel[max_prob_index_Cancel])
```

# Ploting Poisson Distribution

```python
# Make plots
plt.figure(figsize=(12, 5))

# Delay graph
plt.subplot(1, 2, 1)
plt.bar(k_values_delay, prob_delay, color='orange')
plt.title('Probability of Flight Delays in 1 week (Nok Air)')
plt.xlabel('Number of Delayed Flights for 1 week')
plt.ylabel('Probability')

# Cancellations graph
plt.subplot(1, 2, 2)
plt.bar(k_values_cancel, prob_cancel, color='red')
plt.title('Probability of Flight Cancellations in 7 Day (Nok Air)')
plt.xlabel('Number of Canceled Flights for 1 week')
plt.ylabel('Probability')

plt.tight_layout()
plt.show()
```

# Find an Optimal Point

```python
import math

#Delayed flights 
Prob_DelayOp = max_prob_index_Delay / 700
print("Probability of delayed flights:", Prob_DelayOp)

#Canceled flights
Prob_CancelOp = max_prob_index_Cancel / 700
print("Probability of canceled flights:", Prob_CancelOp)

#Find Optimal Point of Delay and Cancel (always round up)
OpDelay = 14 * Prob_DelayOp
RoundDelay = math.ceil(OpDelay)
print("Optimal Point of Delay:", OpDelay)
print("Rounded Optimal Point of Delay (ceil):", RoundDelay)

OpCancel = 14 * Prob_CancelOp
RoundCancel = math.ceil(OpCancel)
print("Optimal Point of Cancel:", OpCancel)
print("Rounded Optimal Point of Cancel (ceil):", RoundCancel)
```
